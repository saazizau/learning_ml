# LIBRARY

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
```

# DATA

```python
df1 = pd.read_csv('decisiontree.csv')
```

```python
df1.describe()
```

```python
df1.info()
```

# PRE-PROCESSING DATA

## DATA TRANSFORMATION

```python
encoding = {
    "mesin": {"bensin": 0, "diesel":1},
    "penggerak": {"depan":0, "belakang":1}
}
df1.replace(encoding, inplace=True)
```

```python
df1
```

Karena sampelnya yang sangat sedikit, maka langsung modeling

```python
df = df1.copy()
```

# MODELING

```python
X = df.drop(['label','ID'], axis='columns')
y = df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

```python
model = tree.DecisionTreeClassifier(criterion='gini', max_depth=5)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```
