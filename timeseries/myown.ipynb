```python
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
```

```python
df = pd.read_csv('TimeSeries.csv')
df.shape
```

```python
dfV = df.head(100)
```

```python
rcParams['figure.figsize'] = 20, 5

plt.plot(dfV.index, dfV['Value'])
plt.xticks(rotation=30)
plt.show()
```
